**First Create Omlette Class Module to Testing and Analyzing ML Model. Create Omlette Class Module to Testing and Analyzing ML Model Can Be Seen As Below.**

**We Can Testing and Analyzing ML Model by using Analyze_and_Testing_ML_Model Function in Automatic Analyze and Testing ML Model OMLETTE Class. Automatic Analyzing and Testing ML Model by using Automatic Analyze and Testing ML Model OMLETTE Class Can be Seen As Below**

**Then We Can See Automatic Analyzing and Testing ML Model Result using Show_Analyze_and_Testing_ML_Result Function in Omlette Class Object. Automatic Analyze and Testing ML Result can be seen as below.**

```python
testing_and_analyzing_ML_model.Show_Analyze_and_Testing_ML_Result()
```

```python
testing_and_analyzing_ML_model.Show_Analyze_and_Testing_ML_Feature_Importance()
```
